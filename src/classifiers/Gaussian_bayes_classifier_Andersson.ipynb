```
# Estimativa de Máxima Verossimilhança P(W_i):
def estMaxVer(class_i, n):
  pwi = 0
  for exemplo_rotulado in range(0,len(class_i)):
    #print(len(class_i))
    #print(exemplo_rotulado)
    pwi += 1
  pwi = pwi / n
  return pwi

# Vetor de médias mi_i para cada classe i pelo conjunto de atributos d
def medias(class_i):
  med_list = []
  N = len(class_i)
  #print("N:",N)
  for n in range(0,d):
    med = 0
    for k in class_i:
      med += X[k-1][n]
    #print(med/N)
    med_list.append(med/N)
  return med_list

# Matriz de Variância-Covariância (dxd)
def funcao(ii):       # Função para criar a partir da variancia^2 a soma dos atributos por classe e dividir por n elementos na classe
  v = []
  for j in range(0,d):  # j está relacionado ao número de d atributos/componentes
    soma = 0
    for pos in particao[ii]:  # Anexar apenas os elementos que pertencem a mesma classe
      soma += (variancia2[ii][pos-1][j])
    v.append(soma/len(particao[ii]))
  return v

# Definir a probabilidade pela normal multivariada. Iterar por cada k (exemplo)
def prod_matriz():
  classe = []
  for i in range(0,C):
    cl = []
    for k in range(0,K):
      resultado = (1 / ((2*math.pi**(d/2)) * (np.linalg.det(covariancia[i])**(1/2)))) * math.exp((-1/2) * variancia[i][k].T.dot(np.linalg.inv(covariancia[i]).dot(variancia[i][k])))
      cl.append(resultado)
    classe.append(cl)
  return classe

def soma_termo_normalizacao(k):
  s = 0
  for r in range(0,C):
    s += prob_normal_multivariada[r][k] * P_wi[r]
  return s
```
